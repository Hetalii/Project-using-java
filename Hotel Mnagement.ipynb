```python
class Member:
    def _init_(self, name, age, gender):
        self.name = name
        self.age = age
        self.gender = gender

class Gym:
    def _init_(self):
        self.members = []

    def add_member(self, name, age, gender):
        member = Member(name, age, gender)
        self.members.append(member)
        print(f"{member.name} has been added to the gym.")

    def remove_member(self, name):
        for member in self.members:
            if member.name.lower() == name.lower():
                self.members.remove(member)
                print(f"{member.name} has been removed from the gym.")
                return
        print("Member not found.")

    def display_members(self):
        print("Gym Members:")
        for member in self.members:
            print(f"Name: {member.name}, Age: {member.age}, Gender: {member.gender}")

    def search_member(self, name):
        for member in self.members:
            if member.name.lower() == name.lower():
                print(f"Name: {member.name}, Age: {member.age}, Gender: {member.gender}")
                return
        print("Member not found.")

    def update_member(self, name, age, gender):
        for member in self.members:
            if member.name.lower() == name.lower():
                member.age = age
                member.gender = gender
                print(f"{member.name}'s information has been updated.")
                return
        print("Member not found.")

    def count_members(self):
        return len(self.members)

    def average_age(self):
        total_age = sum(int(member.age) for member in self.members)
        return total_age / self.count_members() if self.count_members() > 0 else 0

    def oldest_member(self):
        if self.members:
            oldest = max(self.members, key=lambda x: x.age)
            print(f"The oldest member is {oldest.name} aged {oldest.age}.")
        else:
            print("No members in the gym.")

    def youngest_member(self):
        if self.members:
            youngest = min(self.members, key=lambda x: x.age)
            print(f"The youngest member is {youngest.name} aged {youngest.age}.")
        else:
            print("No members in the gym.")

    def gender_distribution(self):
        gender_count = {'Male': 0, 'Female': 0}
        for member in self.members:
            if member.gender.lower() == 'male':
                gender_count['Male'] += 1
            elif member.gender.lower() == 'female':
                gender_count['Female'] += 1
        print("Gender Distribution:")
        for gender, count in gender_count.items():
            print(f"{gender}: {count}")

def main():
    gym = Gym()

    while True:
        print("\n1. Add Member")
        print("2. Remove Member")
        print("3. Display Members")
        print("4. Search Member")
        print("5. Update Member")
        print("6. Count Members")
        print("7. Average Age of Members")
        print("8. Oldest Member")
        print("9. Youngest Member")
        print("10. Gender Distribution")
        print("11. Exit")

        choice = input("Enter your choice: ")

        if choice == '1':
            name = input("Enter member's name: ")
            age = int(input("Enter member's age: "))
            gender = input("Enter member's gender: ")
            gym.add_member(name, age, gender)
        elif choice == '2':
            name = input("Enter member's name to remove: ")
            gym.remove_member(name)
        elif choice == '3':
            gym.display_members()
        elif choice == '4':
            name = input("Enter member's name to search: ")
            gym.search_member(name)
        elif choice == '5':
            name = input("Enter member's name to update: ")
            age = int(input("Enter member's new age: "))
            gender = input("Enter member's new gender: ")
            gym.update_member(name, age, gender)
        elif choice == '6':
            print("Total members:", gym.count_members())
        elif choice == '7':
            print("Average age:", gym.average_age())
        elif choice == '8':
            gym.oldest_member()
        elif choice == '9':
            gym.youngest_member()
        elif choice == '10':
            gym.gender_distribution()
        elif choice == '11':
            print("Exiting program.")
            break
        else:
            print("Invalid choice. Please try again.")

if _name_ == "_main_":
    main()
```
